```python
A = np.array([1,2,3,4])
```

```python
B = np.array([1000,2000,3000,4000])
```

```python
condition = np.array([True, True, False, False])
```

```python
answer = [(a if cond else b) for a,b,cond in zip(A,B,condition)]
```

```python
# A,B,condition の新しいアレイを作る
# condがTrue -> a, False ->b             ->速度が遅い、多次元対応できない
answer #list
```

```python
# 2
answer2 = np.where(condition, A, B)  #->Condition True ->A, False -> B
```

```python
answer2 # array
```
